## **DECISION TREE**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/UNHAS/Datasets/decisiontree_ch6.csv')
```

```python
# Mengganti Kategori Menjadi Kode Angka
encoding = {"mesin":{"bensin":0,"diesel":1},
            "penggerak":{"depan":0, "belakang":1}}
df.replace(encoding, inplace=True)
```

```python
df
```

```python
# Membuang Kolom yang tidak dibutuhkan
X = df.drop(['ID','label'],axis=1)
y = df['label']
```

```python
X  #X adalah feature atau variabel bebas
```

```python
y # y merupakan variabel target
```

```python
# Bagi data menjadi test dan train
import sklearn.model_selection as ms
X_train, X_test, y_train, y_test = ms.train_test_split(X,y, test_size=0.2)
```

```python
# Pembuatan model pertama menggunakan entropy decision tree
import sklearn.tree as tree
model1 = tree.DecisionTreeClassifier(criterion='entropy',max_depth=5)
model1.fit(X_train,y_train)
```

```python
y_prediksi = model1.predict(X_test)
y_prediksi
```

```python
y_test
```

```python
# Metrics untuk melihat akurasi model
import sklearn.metrics as met
```

```python
print(met.accuracy_score(y_test, y_prediksi))
```

```python
# Pembuatan model kedua menggunakan gini decision tree
model2 = tree.DecisionTreeClassifier(criterion='gini',max_depth=5)
model2.fit(X_train,y_train)
```

```python
y_prediksi2 = model2.predict(X_test)
y_prediksi2
```

```python
print(met.accuracy_score(y_test, y_prediksi2))
```

```python
import pydotplus as pp # Melihat Graphic Decision Tree
labels = ['mesin', 'bangku', 'penggerak']
dot_data = tree.export_graphviz(model1, out_file=None, feature_names=labels, filled=True, rounded=True)
graph = pp.graph_from_dot_data(dot_data)
graph.write_png('decisiontree.png')
```

```python
labels = ['mesin', 'bangku', 'penggerak']
dot_data2 = tree.export_graphviz(model2, out_file=None, feature_names=labels, filled=True, rounded=True)
graph2 = pp.graph_from_dot_data(dot_data2)
graph2.write_png('decisiontree2.png')
```
